```python
def get_batches(text, batch_size, seq_length):
    total_length=len(text)
    num_batches=total_length//(batch_size*seq_length)
    margin=seq_length*num_batches
    batches=[]
    for i in range(num_batches):
        batches.append([[],[]])
        for j in range(batch_size):
            idx= i*seq_length + margin*j
            inputs=text[idx:(idx+seq_length)]
            if (idx+seq_length)==total_length:
                targets=inputs[1:]+[text[0]]
            else:
                targets=inputs[1:]+[text[idx+seq_length+1]]
            batches[i][0].append(inputs)
            batches[i][1].append(targets)
    return batches
```

```python
import tensorflow as tf
import numpy as np
```

```python
def get_init_cell(rnn_size, keep_prob=None):
    lstm=tf.contrib.rnn.BasicLSTMCell(num_units=rnn_size)
    #drop=tf.contrib.rnn.DropoutWrapper(lstm, output_keep_prob=keep_prob)
    return lstm
```

```python
rnn_size=512
rnn_layers=2
```

```python
train_graph=tf.Graph()
with train_graph.as_default():
    vocab_size=len(vocab)
    embed_dim=300
    
    inputs=tf.placeholder(dtype=tf.int32, shape=[None, None], name="inputs")
    targets=tf.placeholder(dtype=tf.int32, shape=[None, None], name="targets")
    #learning_rate=tf.placeholder(tf.float32, name="learning_rate")
    #keep_prob=tf.placeholder(tf.float32, name="keep_prob")
    
    inputs_shape=tf.shape(inputs)
    
    embeddings=tf.Variable(tf.truncated_normal(shape=[vocab_size, embed_dim], stddev=1/np.sqrt(vocab_size)))
    embed=tf.nn.embedding_lookup(embeddings, inputs)
    
    cell=tf.contrib.rnn.MultiRNNCell([get_init_cell(rnn_size) for _ in range(rnn_layers)])
    initial_state = cell.zero_state(inputs_shape[0], tf.float32)
    
    outputs, final_state=tf.nn.dynamic_rnn(cell=cell, inputs=embed, dtype=tf.float32, initial_state=initial_state)
    
    #每个单词的分数
    logits=tf.contrib.layers.fully_connected(inputs=outputs, num_outputs=vocab_size, activation_fn=None)
    
    #每个单词的概率
    probs=tf.nn.softmax(logits, name="probs")
    
    #从这里就开始不清楚了
    cost=tf.contrib.seq2seq.sequence_loss(logits, targets, weights=tf.ones(shape=[inputs_shape[0], inputs_shape[1]]))
    
    optimizer=tf.train.AdamOptimizer(0.01)
    
    gradients=optimizer.compute_gradients(cost)
    clipped_gradients=[(tf.clip_by_value(grad,-1.0,1.0),var) for grad,var in gradients if grad is not None]
    train_op=optimizer.apply_gradients(clipped_gradients)
```

```python
epoches=1
batch_size=256
seq_length=5
#lr=0.01
#keep_prob=0.5
show_every_batches=10
save_model_dir="./save"
```

```python
batches=get_batches(ints_words, batch_size, seq_length)
batches=np.array(batches)
```

```python
with tf.Session(graph=train_graph) as sess:
    sess.run(tf.global_variables_initializer())
    #print(batches[0][0])
    #print(keep_prob)
    state=sess.run(initial_state, feed_dict={inputs:batches[0][0]})
    #print(state)
    for i in range(epoches):
        for batch_i,(x,y) in enumerate(batches):
            #print(x)
            train_loss,state, _ =sess.run([cost, final_state, train_op],
                                  feed_dict={inputs:x, targets:y, initial_state:state})
            if batch_i%show_every_batches==0:
                print("epoch:{}, batch:{}/{},loss:{:.3f}".format(i, batch_i, len(batches), train_loss))

    saver=tf.train.Saver()
    saver.save(sess, save_model_dir)
    saver.save(sess, "checkpoints/tv.ckpt")
    print("model trained and saved.")
```

```python
with tf.Session(graph=train_graph) as sess:
    saver=tf.train.Saver()
    saver.restore(sess, "checkpoints/tv.ckpt")
    
    gen_sentences=["吴"]
    gen_length=1000
    prev_state=sess.run(initial_state, feed_dict={inputs:np.array([[1]])})
    #print(prev_state)
    
    for i in range(gen_length):
        dyn_ints=[[word_to_int[word] for word in gen_sentences[-seq_length:]]]
        probabilities,prev_state=sess.run([probs, final_state], feed_dict={inputs:dyn_ints, initial_state:prev_state})
        #print(probabilities)
        #print(np.argmax(probabilities))
        pred_word=int_to_word[np.argmax(probabilities[0][len(dyn_ints)-1])]
        gen_sentences.append(pred_word)
    
    print("".join(gen_sentences))
```
